# Step 1 : Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn import tree
from matplotlib import pyplot as plt
```

# Step 2 : Data Collection, Pre Processing and Analyzing

```python
# Loading the Dataset to DataFrame
df = pd.read_csv('creditcard.csv')
df
```

```python
# First 5 rows of the dataset
df.head()
```

```python
# last 5 rows of the dataset
df.tail()
```

```python
# Dataset Informations
df.info()
```

```python
# Checking the no. of missing values in columns
df.isnull().sum()
```

```python
# Distribution of Legit & Fraudulent Transactions
df['Class'].value_counts()
```

```python
# Dataset is highly unbalanced
# 0 = Legit Transaction
# 1 = Fraud Transaction

# Seperating the data for analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measure of data
legit.Amount.describe()
```

```python
# Statistical measure of data
fraud.Amount.describe()
```

```python
# Compare values for both transactions
df.groupby('Class').mean()
```

```python
# Under Sampling
# Build a sample dataset containing similar distribution of legit and fraud transaction
# No. of Fraud Transaction = 492
legit_sample = legit.sample(n=492)
```

```python
# Concatenating two DataFrames
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# Step 3 : Train and Test Splitting

```python
# Splitting the data into Features & Targets
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into Training & Testing Data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Step 4 : Logistic Regression Model

```python
# Model Training
lg_model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data
lg_model.fit(X_train, Y_train)
```

# Step 5 : Logistic Regression Model Evaluation

```python
# Accuracy Score
# Accuracy on Training Data
X_train_prediction = lg_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data :', training_data_accuracy)
```

```python
# Accuracy on Test Data
X_test_prediction = lg_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data : ', test_data_accuracy)
```

# Step 6 : SVC Model

```python
# Model Training
svm_model = svm.SVC(kernel="linear", gamma="auto", C=2)
```

```python
# Training the SVM Model with Training Data
svm_model.fit(X_train, Y_train)
```

# Step 7 : SVC Model Evaluation

```python
# Accuracy Score
# Accuracy on Training Data
X_train_prediction = svm_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data :', training_data_accuracy)
```

```python
# Accuracy on Test Data
X_test_prediction = svm_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data : ', test_data_accuracy)
```

# Step 8 : Decision Tree Classifier Model

```python
# Model Training
dtc_model = DecisionTreeClassifier(max_depth=3)
```

```python
# Training the Decision Tree Classifier Model with Training Data
dtc_model.fit(X_train, Y_train)
```

```python
fig = plt.figure()
```

```python
tree.plot_tree(dtc_model)
plt.show()
```

# Step 9 : Decision Tree Classifier Model Evaluation

```python
# Accuracy Score
# Accuracy on Training Data
X_train_prediction = dtc_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data :', training_data_accuracy)
```

```python
# Accuracy on Test Data
X_test_prediction = dtc_model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data : ', test_data_accuracy)
```
